# Discussion Questions

### Describe your reaction to the match rate. What sources of data might be prone to privacy issues?

The high match rate is concerning as it demonstrates how easily anonymized data can be re-identified using simple matching techniques. This shows that removing direct identifiers like names is insufficient for privacy protection when quasi-identifiers remain. Data sources particularly prone to privacy issues include medical records (when combined with voter registration data), location tracking data, and educational records

### Provide examples of at least two other variables that could increase the match rate if included.

Two additional variables that would increase the match rate are occupation and education level. Occupation helps distinguish between individuals with identical demographics since people in the same age group, gender, and zip code  work in different fields. Education level also adds uniqueness because even within the same geographic area, educational attainment varies. These variables would create more distinctive identifier combinations, making it more likely that anonymized records match to exactly one person in the auxiliary dataset
